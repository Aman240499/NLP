```python
#import libraries
import spacy
import pandas as pd
from spacy import displacy
import matplotlib.pyplot as plt
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.pipe_names
```

```python
#example data
doc = 'Mr.Elon Musk acquired Twitter Organization for $45 billion in 2022'

ner = nlp(doc)
ner
```

```python
#entity in given sentence
ner.ents
```

```python
for i in ner.ents:
  print(i,":",i.label_ )
```
